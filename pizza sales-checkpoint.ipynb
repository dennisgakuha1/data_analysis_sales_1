---**Daily trend for total orders**

```python
# SET SQL_SAFE_UPDATES = 0;

# UPDATE pizza_sales
# SET order_date = STR_TO_DATE(order_date, '%d-%m-%Y');

# ALTER TABLE pizza_sales
# MODIFY order_date DATE;
```

```python
%%sql
SELECT
    DAYNAME(order_date) AS day_of_week,
    COUNT(DISTINCT order_id) AS no_of_orders
FROM pizza_sales
GROUP BY day_of_week
ORDER BY no_of_orders DESC;
```

```python
%%sql
SELECT
    DAYOFWEEK(order_date) AS order_day,
    DAYNAME(order_date) AS day_of_week,
    COUNT(DISTINCT order_id) AS no_of_orders
FROM pizza_sales
GROUP BY
    order_day,
    day_of_week
ORDER BY
    order_day;
```

```python
%%sql
SELECT
    DAYNAME(order_date) AS day_of_week,
    COUNT(DISTINCT order_id) AS no_of_orders
FROM pizza_sales
GROUP BY
    DAYOFWEEK(order_date),
    DAYNAME(order_date)
ORDER BY
    DAYOFWEEK(order_date);

```

```python
%%sql
SELECT
    DAYNAME(order_date) AS day_of_week,
    COUNT(DISTINCT order_id) AS no_of_orders
FROM pizza_sales
GROUP BY
    WEEKDAY(order_date),
    DAYNAME(order_date)
ORDER BY
    WEEKDAY(order_date);

# STARTS WITH MON
```

```python
%%sql
SELECT
    DATE(order_date) AS order_day,
    DAYNAME(order_date) AS day_of_week,
    COUNT(DISTINCT order_id) AS no_of_orders
FROM pizza_sales
GROUP BY
    DATE(order_date),
    DAYNAME(order_date)
ORDER BY
    order_day
limit 5;
```

---**Monthly trend for total orders**

```python
%%sql
SELECT
    MONTHNAME(order_date) AS month,
    COUNT(DISTINCT order_id) AS no_of_orders
FROM pizza_sales
GROUP BY month
ORDER BY no_of_orders DESC;
```

```python
%%sql
SELECT
    MONTH(order_date) AS month_no,
    MONTHNAME(order_date) AS month_name,
    COUNT(DISTINCT order_id) AS no_of_orders
FROM pizza_sales
GROUP BY
    MONTH(order_date),
    MONTHNAME(order_date)
ORDER BY
    month_no;
```

```python
%%sql
SELECT
    YEAR(order_date) AS year,
    MONTH(order_date) AS month,
    COUNT(*) AS total_orders
FROM pizza_sales
GROUP BY YEAR(order_date), MONTH(order_date)
ORDER BY year, month;
```

---**Hourly trend for total orders**

```python
# SET SQL_SAFE_UPDATES = 0;

# UPDATE pizza_sales
# SET order_time = STR_TO_DATE(order_time, '%H:%i:%s');

# ALTER TABLE pizza_sales
# MODIFY order_time TIME;
```

```python
%%sql
SELECT
    HOUR(order_time) AS hour,
    COUNT(DISTINCT order_id) AS no_of_orders
FROM pizza_sales
GROUP BY hour
ORDER BY no_of_orders DESC;
```

```python
%%sql
SELECT
    HOUR(order_time) AS order_hour,
    COUNT(DISTINCT order_id) AS no_of_orders
FROM pizza_sales
GROUP BY order_hour
ORDER BY order_hour;
```

**--%sales by pizza category**

```python
%%sql
SELECT 
pizza_category,
cast(sum(total_price)AS DECIMAL(10,2)) as revenue,
cast(sum(total_price)*100/
(select sum(total_price)
from pizza_sales)
AS DECIMAL(10,2)
) as pct

from pizza_sales
group by pizza_category
order by pct DESC;
```

**--%sales by pizza size**

```python
%%sql
SELECT 
pizza_size,
cast(sum(total_price)AS DECIMAL(10,2)) as revenue,
cast(sum(total_price)*100/
(select sum(total_price)
from pizza_sales)
AS DECIMAL(10,2)
) as pct

from pizza_sales
group by pizza_size
order by pct DESC;
```

**-- total pizzas sold per category**

```python
%%sql
SELECT 
pizza_category,
sum(quantity) as quantity_sold
from pizza_sales
group by pizza_category
order by quantity_sold DESC;
```

**---Top 5 Pizzas by Revenue**

```python
%%sql
SELECT 
pizza_name,
cast(sum(total_price) as decimal(10,2)) as revenue
from pizza_sales
group by pizza_name
order by revenue DESC
limit 5;
```

**---Bottom 5 Pizzas by Revenue**

```python
%%sql
SELECT 
pizza_name,
cast(sum(total_price) as decimal(10,2)) as revenue
from pizza_sales
group by pizza_name
order by revenue ASC
limit 5;
```

**---Top 5 Pizzas by quantity**

```python
%%sql
SELECT 
pizza_name,
sum(quantity) as quantity_sold
from pizza_sales
group by pizza_name
order by quantity_sold DESC
limit 5;
```

**---Bottom 5 Pizzas by quantity**

```python
%%sql
SELECT 
pizza_name,
sum(quantity) as quantity_sold
from pizza_sales
group by pizza_name
order by quantity_sold ASC
limit 5;
```
